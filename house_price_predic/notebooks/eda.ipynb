

# Exploratory Data Analysis (EDA)

This notebook explores the California Housing dataset and analyzes relationships between features and house prices.

Target Distribution

```python
plt.figure(figsize=(8,5))

df["Price"].hist(bins=30)

plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

### Observation

Most house prices are concentrated between 1 and 3. The distribution is right-skewed, with fewer high-priced houses. A spike is visible near the maximum value due to price capping in the dataset.

Correlation Heatmap

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

Median Income has the strongest positive correlation with house price (0.69), indicating that higher-income areas generally have higher house prices.

Scatter plot 1

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="MedInc",
    y="Price",
    data=df
)

plt.title("Median Income vs House Price")

plt.show()
```

### Observation

House prices tend to increase as median income increases, showing a strong positive relationship between income and housing prices.

Scatter plot 2

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="AveRooms",
    y="Price",
    data=df
)

plt.title("Average Rooms vs House Price")

plt.show()
```

### Observation

Average rooms show a weak positive relationship with house prices, and several outliers are visible in the dataset.
